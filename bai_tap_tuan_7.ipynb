```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LinearRegression
```

```python
url='/content/passenger.csv'
data= pd.read_csv(url)
data.head()
```

```python

data.hist()
plt.show()
```

```python
correlation=data.corr()
print(correlation)
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(data.corr(), annot=True)
```

```python
model = LinearRegression()

model.fit(X=pd.DataFrame(data['Miles(X)']),y= pd.DataFrame(data['Dollars (Y)']))

print('Intercept: ', model.intercept_)
print('Coeffient: ', model.coef_)
print('Độ chính xác của mô hình : ', model.score(X=pd.DataFrame(data['Miles(X)']),y= pd.DataFrame(data['Dollars (Y)'])))

data.plot(kind='scatter', x='Miles(X)', y='Dollars (Y)', figsize=(9, 9), color='Blue', xlim=(0, 6000))

plt.plot(pd.DataFrame(data['Miles(X)']),model.predict(pd.DataFrame(data['Miles(X)'])) , color='black')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LinearRegression
```

```python
url='/classification.csv'
data= pd.read_csv(url)
data.head()
```

```python
data.hist()
plt.show()
```

```python
correlation=data.corr()
print(correlation)
```

```python
sns.pairplot(data)
```

```python
model = LinearRegression()
model.fit(X=pd.DataFrame(data['interest']), y=data['success'])

X = pd.DataFrame(data['interest'])

print('Intercept: ', model.intercept_)
print('Coeffient: ',model.coef_)
print('Độ chính xác của mô hình : ', model.score(X, y=data['success']))

y_pred = model.predict(X=pd.DataFrame(data['interest']))
```

```python
model = LinearRegression()

X = pd.DataFrame([data['interest'], data['interest']**2, data['interest']**3, data['interest']**4, data['interest']**5]).T

model.fit(X, y=data['success'])

print('Intercept: ', model.intercept_)
print('Coeffient: ', model.coef_)
print('Độ chính xác của mô hình : ', model.score(X, y=data['success']))

range = np.arange(0, 101, 1)
X_range = pd.DataFrame([range, range**2, range**3, range**4, range**5]).T

y_pred = model.predict(X=X_range)
data.plot(kind='scatter', x='interest', y='success', figsize=(9, 9), color='blue', xlim=(0, 101))
plt.plot(range, y_pred, color='black')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LinearRegression
```

```python
url='/Position_Salaries.csv'
data= pd.read_csv(url)
data.head()
```

```python
data.hist()
plt.show()
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(data.corr(), annot=True)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X=pd.DataFrame(data['Level']), y=data['Salary'])
X = pd.DataFrame(data['Level'])
y=data['Salary']
print('Intercept: ', model.intercept_)
print('Coeffient: ', model.coef_)
print('Độ chính xác của mô hình : ', model.score(X, y=data['Salary']))

y_pred = model.predict(X=pd.DataFrame(data['Level']))
data.plot(kind='scatter', x='Level', y='Salary', figsize=(9, 9), color='blue', xlim=(0, 11), ylim=(0, 1500000))
plt.plot(data['Level'], y_pred, color='black')
```

```python
model = LinearRegression()

X = pd.DataFrame([data['Level'], data['Level']**2]).T

model.fit(X, y=data['Salary'])

print('Intercept: ', model.intercept_)
print('Coeffient: ', model.coef_)
print('Độ chính xác: ', model.score(X, y=data['Salary']))

range = np.arange(0, 11, 0.1)
X_range = pd.DataFrame([range, range**2]).T

y_pred = model.predict(X=X_range)
data.plot(kind='scatter', x='Level', y='Salary', figsize=(9, 9), color='blue', xlim=(0, 11))
plt.plot(range, y_pred, color='black')
```

```python
model = LinearRegression()

X = pd.DataFrame([data['Level'], data['Level']**2,data['Level']**3,data['Level']**4,data['Level']**5,data['Level']**6,data['Level']**7,data['Level']**8,data['Level']**9,data['Level']**10]).T

model.fit(X, y=data['Salary'])

print('Intercept: ', model.intercept_)
print('Coeffient: ', model.coef_)
print('Độ chính xác: ', model.score(X, y=data['Salary']))

range = np.arange(0, 11, 0.1)
X_range = pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T
y_pred = model.predict(pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T)
data.plot(kind='scatter', x='Level', y='Salary', figsize=(9, 9), color='blue', xlim=(0, 11))
plt.plot(range, y_pred, color='black')
```
